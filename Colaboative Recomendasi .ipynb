# Collaborative Recomendation

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('anime.csv')
df2 = pd.read_csv('rating.csv')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2['rating'] = df2['rating'].replace(-1,0)
# # df2 = df2.dropna()
# df2.isnull().sum()
# df2.info()
```

```python
## File kegedean, kita filter
df2 = df2.iloc[:10000]
df2
```

```python
df3 = df2.pivot_table(columns=['anime_id'],index=['user_id'],values=['rating'])
df3 = df3.fillna(0)
df3
```

```python
dfcor = df3.corr()
dfcor = df3.fillna(0)
dfcor.head(10)
```

## Rekomendasi Ali

```python
## Misal Ali nonton anime 1:10
Ali = [1,10]
skor = dfcor['rating'][Ali[0]] * Ali[1]
skor = skor.sort_values(ascending=False)
# skor[:20]
```

```python
anime_id = []
for i in range(10):
    anime_id.append(skor.index[i][1])
anime_id
```

```python
index = []
for i in anime_id:
    index.append(df[df['anime_id']==i].index[0])
dfAli = df.iloc[index]
dfAli
```

## Rekomendasi Bebi

```python
# Misal Bebi nonotn film 6 dan 15
Bebi = [[6,9],[15,9]]
```

```python
dfSkor = pd.DataFrame()
for produk, rating in Bebi:
    skor = dfcor['rating'][produk] *  rating
    dfSkor = dfSkor.append(skor)
dfSkor
```

```python
# rekomendasi
rek_Bebi = dfSkor.sum().sort_values(ascending=False)
rek_Bebi[:10]
```

```python
anime_id = []
for i in range(10):
    anime_id.append(rek_Bebi.index[i][1])
anime_id
```

```python
index = []
for i in anime_id:
    index.append(df[df['anime_id']==i].index[0])
dfBebi = df.iloc[index]
dfBebi
```

```python
index = []
for i in anime_id:
    index.append(df[df['anime_id']==i].index[0])
dfCindy = df.iloc[index]
dfCindy
```

## User ID 5

```python
anime_id = df2['anime_id'][df2['user_id']==5].tolist()
rating = df2['rating'][df2['user_id']==5].tolist()
anime_id[:10]
```

```python
Dodi = list(map(lambda i,j: (i,j), anime_id,rating))
Dodi = sorted(Dodi, key=lambda x: x[1], reverse=True)
Dodi[:10]
```

```python
dfSkor = pd.DataFrame()
for produk, rating in Dodi:
    skor = dfcor['rating'][produk] *  rating
    dfSkor = dfSkor.append(skor)
dfSkor
```

```python
rek_Dodi = dfSkor.sum().sort_values(ascending=False)
anime_id = []
for i in range(10):
    anime_id.append(rek_Dodi.index[i][1])
index = []
for i in anime_id:
    index.append(df[df['anime_id']==i].index[0])
dfDodi = df.iloc[index]
dfDodi
```
