__B. Count Vectorizer__

- bagaimana menghitung cos similarity untuk categorical data (value berupa string)
- Data di-extract untuk menentukan jumlah tiap kata. Jumlah tiap kata dalam data itulah yang dicari cosine similarity-nya
- di sklearn, gunakan ```CountVectorizer```

#### ```from sklearn.feature_extraction.text import CountVectorizer```

````python
from sklearn.feature_extraction.text import CountVectorizer
````

````python
x = 'Andi Budi Caca'
y = 'Andi Budi Caca'
````

````python
countvec = CountVectorizer()
````

````python
c = countvec.fit_transform([x, y])
c
````

````python
#kata yg terkandung
countvec.get_feature_names()
````

````python
# matriks frequency : freq tiap kata di setiap data
mf = c.toarray()
mf

#[[1, 1, 1] => data 1 = andi 1x, budi 1x, caca 1x
#[1, 1, 1]] => data 2 = andi 1x, budi 1x, caca 1x
````

````python
similarityScore = cosine_similarity(mf)
similarityScore

#       X  Y
#  X [[1., 1.]
#  Y  [1., 1.]]
````

### _CONTOH_

- Content-Based Filtering by __*Brand*__ and __*Color*__

````python
df = pd.read_csv('USA_cars_datasets.csv')
df.head()
````

````python
# create column contains value combinations between brand and color

df['BC'] = df.apply(lambda row: str(row['brand']) + ' ' + str(row['color']), axis=1)
df['MC'] = df['model'] + ' ' +  df['color']
df.head(3)

df[df['BC'] == 'toyota black']
````

````python
# count vectorizer : hitung kata di tiap data di column BC

cv = CountVectorizer()
cm = cv.fit_transform(df['BC'])

# total kata unik di column BC
print(cv.get_feature_names())

# matrix frequency tiap kata 
print(cm.toarray())
print(cm.toarray()[0])
````

````python
# hitung cosine similarity dari tiap data di matrix frequency

similarityScore = cosine_similarity(cm.toarray())
similarityScore[0]
````

````python
# gunakan similarity score sebagai data rekomendasi
# misal saya suka mobil data pertama (index ke-0)

# daftar seluruh mobil beserta similarity scorenya
similarCars = list(enumerate(similarityScore[0]))
similarCars # (index, nilai kesamaannya)
````

````python
# sort berdasarkan similarity-nya
similarCars = sorted(similarCars, key=lambda x: x[1], reverse=True)
similarCars
````

````python
# 5 mobil yang mirip

similarCars[:6]
````
